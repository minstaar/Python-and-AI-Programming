```python
import pandas as pd

score = [["A", 70, 80, 100],
         ["B", 100, 80, 70],
         ["C", 90, 95, 90]]

score_df = pd.DataFrame(score)
score_df
```

```python
score_df.columns = ["Name", "Subject1", "Subject2", "Subject3"]
score_df
```

```python
score_df_r = score_df.rename(columns={"Subject1":"Math", "Subject2":"English", "Subject3":"Science"})
score_df_r
```

```python
score_df_r.values
```

```python
print(score_df_r.shape)
score_df_r.dtypes
```

```python
score_dict = {'Name':['A', 'B', 'C'], 'Subject1':[70, 100, 90], 'Subject2':[80, 80, 95], 'Subject3':[100, 70, 90]}
print(type(score_dict))

score_dict_df = pd.DataFrame(score_dict)
score_dict_df
```

```python
print(score_dict_df.index)
print(score_dict_df.columns)
print(score_dict_df.values)
```

```python
print(score_dict_df.dtypes)

score_dict_ndf = score_dict_df.astype({'Subject1':float, 'Subject2':float, 'Subject3':float})
print(score_dict_ndf)
print(score_dict_ndf.dtypes)
```

```python
import pandas as pd

x = [0, 1, 2, 3, 4, 5]
y = [i * 2 for i in x]

df = pd.DataFrame({'x_value':x, 'y_value':y})
df
```

```python
df['xy_value'] = df['x_value'] * df['y_value']
df
```

```python
index = ['zero', 'one', 'two', 'three', 'four', 'five']
df['number'] = index
df
```

```python
df.set_index('number', append = True)
df
```

```python
df2 = df.set_index('number', append = False)
df2
```

```python
df3 = df2.reset_index(drop = True)
df3
```

```python
dates = pd.date_range('20240101', periods = 6)
dates
```

```python
df4 = df3.set_index(dates)
df4
```

```python
print(df4.index)
print(df4.columns)
print(df4.values)
```

```python
print(df2.plot())
```

```python
df4.sort_values(by = 'y_value', ascending = False)
```

```python
df4['mean'] = df4.mean(axis = 1)
df4
```
